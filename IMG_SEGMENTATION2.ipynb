```python
import numpy as np
import cv2
```

```python
img = cv2.imread('NORMAL.png')
print(img.shape)
```

```python
# flatten the img

img2 = img.reshape((-1, 3))

print(img2.shape)
```

```python
img3 = np.float32(img2)

citeria = (cv2.TERM_CRITERIA_EPS+cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)

#Clusters

k=4

attempts = 10


ret, label, center = cv2.kmeans(img3, k, None, citeria, attempts, cv2.KMEANS_RANDOM_CENTERS)
```

```python
center = np.uint8(center)

res = center[label.flatten()]

res2 = res.reshape((img.shape))
```

```python
cv2.imwrite('Seg.jpg', res2)
```

```python
import matplotlib.pyplot as plt

img = cv2.imread('Seg.jpg')
plt.imshow(img, cmap='gray')
plt.show()

img01 = cv2.imread('NORMAL.png')
plt.imshow(img01, cmap='gray')
plt.show()
```

```python
img = cv2.imread('yara.jpg')
print(img.shape)

# flatten the img

img2 = img.reshape((-1, 3))
print(img2.shape)


img3 = np.float32(img2)

citeria = (cv2.TERM_CRITERIA_EPS+cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)

#Clusters

k=6

attempts = 10


ret, label, center = cv2.kmeans(img3, k, None, citeria, attempts, cv2.KMEANS_RANDOM_CENTERS)

center = np.uint8(center)

res = center[label.flatten()]

res2 = res.reshape((img.shape))

cv2.imwrite('Segyara.jpg', res2)
```

```python
import matplotlib.pyplot as plt

img = cv2.imread('Segyara.jpg')
plt.imshow(img, cmap='gray')
plt.show()

img01 = cv2.imread('yara.jpg')
plt.imshow(img01, cmap='gray')
plt.show()
```
